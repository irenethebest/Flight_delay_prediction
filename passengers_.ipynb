```python
import pandas as pd
import numpy as np
```

```python
flights = pd.read_csv('flights_2019_weather_full.csv')
```

```python
passengers = pd.read_csv('average passengers.csv')
```

```python
flights['month'] = pd.DatetimeIndex(flights['fl_date']).month
flights['year'] = pd.DatetimeIndex(flights['fl_date']).year
```

```python
passengers.head()
```

```python
flights.head()
```

```python
flights_pass = flights.merge(passengers,how='left',on=['month','origin','dest'] )
```

```python
flights_pass['avg_passenger_capacity_usage'] = flights_pass['average_passengers']/flights_pass['average_seats']
```
