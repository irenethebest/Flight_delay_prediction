```python
# fuel_consultion includes international flights. to exclude those international flights
# Q. do we need to use fuel_consumption data??
```

```python
fuel_consumption['sdomt_gallons'].value_counts()
```

```python
fuel_consumption= fuel_consumption.drop(fuel_consumption[(fuel_consumption.tdomt_cost == 0) | (fuel_consumption.tint_cost > 0)].index)
fuel_consumption
```

```python
fuel_consumption = fuel_consumption.dropna(how='any',axis=0) 
fuel_consumption
```

```python
fuel_consumption = fuel_consumption.drop(['airline_id', 'carrier', 'carrier_name','satl_gallons', 
                                          'spac_gallons','slat_gallons', 'sint_gallons', 'ts_gallons',
                                          'satl_cost', 'spac_cost','slat_cost', 'sint_cost'], axis=1)
```

```python
#tdomt_gallons + tint_gallons(0) = total_gallons -> to delete domt_gallons, tint_gallons
#sdomt_cost = ts_cost -> to delete sdomt_cost
#total_cost = tdomt_cost + tint_cost(0) -> to delete tdomt_cost and tinit_cost
fuel_consumption = fuel_consumption.drop(['tint_gallons','tdomt_gallons','sdomt_cost','tint_cost',
                                          'tdomt_cost','sdomt_gallons','ts_cost'], axis=1)
fuel_consumption.reset_index(drop=True)
```

```python
fuel_consumption.to_csv('fuel_fe.csv', index=False)
```
