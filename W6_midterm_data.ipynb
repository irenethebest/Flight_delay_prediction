```python
y = flights['arr_delay']
X = flights.select_dtypes(exclude=['object'])
X = flights.drop('arr_delay', axis=1)
```

```python
type(y)
```

```python
# Importing required libraries
import matplotlib.pyplot as plt
import numpy as np

#Creating a Function.
def normal_dist(y , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((y-mean)/sd)**2)
    return prob_density
 
#Calculate mean and Standard deviation.
mean = y.mean()
sd = y.std()
 
#Apply function to the data.
pdf = normal_dist(y,mean,sd)
 
#Plotting the Results
plt.plot(y,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```
