```python
flights = flights.drop(['branded_code_share','cancelled','cancellation_code','diverted','dup','first_dep_time', 'total_add_gtime',
       'longest_add_gtime', 'no_name'], axis=1)
flights_delay = flights[['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay',
       'late_aircraft_delay']]
flights = flights.drop(['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay',
       'late_aircraft_delay'], axis=1)
flights.info()
```

```python
flights = flights.dropna(how='any',axis=0) 
flights.head()
```

```python
flights.columns
```

```python
flight_time = flights[[ 'crs_dep_time', 'dep_time',
       'dep_delay', 'taxi_out', 'wheels_off', 'wheels_on', 'taxi_in',
       'crs_arr_time', 'arr_time', 'arr_delay', 'crs_elapsed_time',
       'actual_elapsed_time', 'air_time']]
```

```python
# flights dep_time + taxi_out = wheels_off 
# wheels_on + taxi_in = arr_time
# 
flights = flights.drop(['mkt_carrier', 'mkt_carrier_fl_num','op_unique_carrier', 'tail_num', 
                        'op_carrier_fl_num','origin_airport_id','origin_city_name', 
                        'dest_airport_id','dest_city_name','dep_time', 'dep_delay',
       'taxi_out', 'wheels_off', 'wheels_on', 'taxi_in','flights'], axis=1)
flights
```

```python
import datetime as dt
flights['month'] = pd.to_datetime(flights['fl_date']).dt.month
```

```python
flights.to_csv('flights_fe.csv', index=False)
```
